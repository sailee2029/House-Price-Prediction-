```python
#Importing Libraries

import pandas as pd
import csv
```

```python
#Giving DataSet Path

DataPath="C:\\Users\\Sailee\\Downloads\\Banglore_Data.csv"
```

```python
#Reading Data

ReadData=pd.read_csv(DataPath)
print("Data is Read...")
```

```python
#Extracting Features from ReadData

Features=['total_sqft','bath','balcony']
X=ReadData[Features] #independent variable 
y=ReadData['price'] #dependent variable 
```

```python
#Splitting Data As Testing and Training

from sklearn.model_selection import train_test_split 

train_X,Val_X,train_y,Val_y=train_test_split(X,y,random_state=1)

print("Data is Splitted into Training and Testing Frames...")
```

```python
#Building Linear Regression Model>>>>>>

from sklearn.linear_model import LinearRegression 

Model=LinearRegression() #calling regression 

Model.fit(train_X,train_y) # training model with training data

evaluate_1=Model.score(train_X,train_y) #Evaluating trained model with R square Score technique

print("R square of Training Data is : ",evaluate_1)

#Testing Model with Testing Data and Getting out Predictions>>

Predict=Model.predict(Val_X)

evaluate_2=Model.score(Val_X,Val_y) #evaluating model after building by test data

print("R square of Testing Data is : ",evaluate_2)
"""
store_as_csv=pd.DataFrame({"ID":Val_y.index,"Validation Data":Val_y,"Predicted Data":Predict})
store_as_csv.to_csv("Model Prediction Data.csv",index=False)
"""
print("Done Successfully!!\n\n")
```

```python
#Now Graph to be plotted here using matplotlib>>>>>>>>>>>>>>>>>>>>

import matplotlib.pyplot as plt
#preprocessing for graph plotting

V_y=sorted(Val_y)

Predict=[round(i,3) for i in Predict]#making data uniform
Predict=sorted(Predict)#sorting for better linear graph
```

```python
#Graph 1 : total_sqft V/s V_y & Predict
V_x_totalsqft=Val_X['total_sqft']
V_x_totalsqft=sorted(V_x_totalsqft)#sorting for better linear graph
bathh=Val_X['bath']
bathh=sorted(bathh)
balconyy=Val_X['balcony']
balconyy=sorted(balconyy)

```

```python
plt.plot(V_x_totalsqft,V_y, label="Actual Price")
plt.plot(V_x_totalsqft,Predict, label="Predicted Price")
plt.title('Total Area Sqft v/s Price')
plt.xlabel('Total area Sqft')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
plt.plot(bathh,V_y,label="Actual Price")
plt.plot(bathh,Predict, label="Predicted Price")
plt.title('No. of Bathroom  v/s Price')
plt.xlabel('Bathroom')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
plt.plot(balconyy ,V_y,label="actual price")
plt.plot(balconyy,Predict,label="predicted price")
plt.title('No.of balconies v/s Price')
plt.xlabel('Balcony')
plt.ylabel('Price')
plt.legend()
plt.show()
```
